IMPORTANDO AS BIBLIOTECAS QUE SERÃO UTILIZADAS

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

EXEMPLO DO PASSO A PASSO DO K-MEANS

Criando um conjunto de dados

```python
dataset = np.array([[1,4],[2,2],[2,5],[3,3],[3,5],[4,7],[5,6],[6,4],[6,7],[7,6],[7,9],[8,7],[8,9],[9,4],[9,8]])

# Plotando os pontos [x,y] num gráfico de dispersão
plt.scatter(dataset[:,0],dataset[:,1])
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()

# Executando a apresentação do(s) gráfico(s) em tela
plt.show()
```

```python
# Criando um agrupamento (algoritmo) otimizado para 3 clusters com 300 iterações máximas (critério de parada)
kmeans = KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10)

# Ajustando e criando o modelo conforme os pontos [x,y]
pred_y = kmeans.fit_predict(dataset)

# Plotando os pontos [x,y]
plt.scatter(dataset[:,0],dataset[:,1])
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()

# Plotando os centroides criados pelo K-Means
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=70,c='red')

# Executando a apresentação do(s) gráfico(s) em tela
plt.show()
```

```python
# Plotando os pontos [x,y] com uma cor correspondente ao cluster definido na etapa de ajuste
plt.scatter(dataset[:,0],dataset[:,1],c=pred_y)
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()

# Plotando os centroides criados pelo K-Means
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=70,c='red')

# Executando a apresentação do(s) gráfico(s) em tela
plt.show()
```

DADOS ALEATORIOS

```python
# Criando 500 amostras "aleatórias" (aleatoriedade iniciando a partir de um mesmo ponto) com 50 centroides (clusters) definidos aleatoriamente
X,Y = make_blobs(n_samples=500,centers=50,random_state=0)

# Plotando os pontos [x,y] gerados aleatoriamente
plt.scatter(X[:,0],X[:,1])

# Executando a apresentação do(s) gráfico(s) em tela
plt.show()
```

```python
# Criando uma lista com todas as somas dos quadrados das distâncias para o centroides correspondente para o intervalo entre 1 a 20 clusters
cluster_minimo = 1
cluster_maximo = 20
wcss = []
for i in range(cluster_minimo, cluster_maximo):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) #SOMA DO QUADRADO DAS DISTANCIAS PARA TODOS OS CENTROIDES
  #METODO CUTOVELO DA CURVA

# Apresentando esses valores (y) para cada número de clusters (x) em gráfico de linha
plt.plot(range(cluster_minimo, cluster_maximo), wcss)

# Apresentando esses valores (y) para cada número de clusters (x) em gráfico de linha# Apresentando esses valores (y) para cada número de clusters (x) em gráfico de dispersão
plt.scatter(range(cluster_minimo, cluster_maximo), wcss)

plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')

# Executando a apresentação do(s) gráfico(s) em tela
plt.show()

# É possível perceber visualmente que a forma do cotovelo se inicia a partir do 4º cluster, ou seja, o decrescimento das somas começa a se tornar não significativo a partir deste cluster
```

```python
# Criando um agrupamento para 4 clusters
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)

# Ajustando o modelo conforme os pontos
pred_y = kmeans.fit_predict(X)

# Plotando os pontos [x,y] com uma cor correspondente ao cluster definido na etapa de ajuste
plt.scatter(X[:,0], X[:,1], c=pred_y)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=50, c='red')

# Executando a apresentação do(s) gráfico(s) em tela
plt.show()
```
